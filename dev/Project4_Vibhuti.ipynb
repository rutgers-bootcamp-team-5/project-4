```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler


import tensorflow as tf
```

```python
# Import and read the alzheimers_disease_data.csv.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/rutgers-bootcamp-team-5/project-4/main/data/alzheimers_disease_data.csv")
df.head()
```

```python
# Seperate the features, X,  from the target variable, y
y = df['Diagnosis']
X = df.drop(columns='Diagnosis')
```

```python
# Preview the features data
X.head()
```

```python
# Preview the first twenty entries for the target variable
y[:20]
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
```

```python
# Review the features data
X.head()
```

```python
# Split the dataset using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Instantiate a StandardScaler instance
scaler = StandardScaler()

# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNeighborsClassifier model with n_neighbors = 3
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
```

```python
KNeighborsClassifier(n_neighbors=3)
```

```python
# Create predictions using the testing data
y_pred = knn.predict(X_test_scaled)
```

```python
# Print the classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred))
```

```python
## Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
number_input_features = len(X_train_scaled[0])
nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=4, input_dim=number_input_features, activation="relu"))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=2,activation='relu'))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
